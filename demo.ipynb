```python
from worderivative import WorDerivative
```

### Specify the WorDerivative. 

```python
der = WorDerivative("that", mid_price=3500, price_range=50, letter_height_scaling=500, call_pricing_strike_thresholds=[2000, 4000], call_pricing_degree_of_approx=2)
```

### Display meta information, describing to the provided underlying asset and its option data. 

```python
der.call.meta
```

### Display the fit of the call price approximation to your sample data.

```python
der.call.plot_price_chart()
```

### Track the the journey of your input word along its digitized version, its translation into letter-shaped derivative payoffs, and, ultimately, its aggregated payoffs. 

```python
der.plot_word_matrix()
```

```python
der.plot_payoff(x_boundaries=[3450, 3550], aggregated=False)
```

```python
der.plot_payoff(x_boundaries=[3450, 3550], aggregated=True)
```
